```python
#label encoding object data typed columns

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
data['Year'] = label_encoder.fit_transform(data['Year'])
data['Season'] = label_encoder.fit_transform(data['Season'])
data['Koppen climate classification'] = label_encoder.fit_transform(data['Koppen climate classification'])
data['Climate'] = label_encoder.fit_transform(data['Climate'])
data['City'] = label_encoder.fit_transform(data['City'])
data['Country'] = label_encoder.fit_transform(data['Country'])
data['Cooling startegy_building level'] = label_encoder.fit_transform(data['Cooling startegy_building level'])

data
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Fill missing values with mean or mode

data['Thermal sensation'] = data['Thermal sensation'].fillna(data['Thermal sensation'].mean())
data['PMV'] = data['PMV'].fillna(data['PMV'].mean())
data['PPD'] = data['PPD'].fillna(data['PPD'].mean())
data['SET'] = data['SET'].fillna(data['SET'].mean())
data['Met'] = data['Met'].fillna(data['Met'].mean())
data['Clo'] = data['Clo'].fillna(data['Clo'].mean())
data['Air temperature (C)'] = data['Air temperature (C)'].fillna(data['Air temperature (C)'].mean())
data['Ta_h (C)'] = data['Ta_h (C)'].fillna(data['Ta_h (C)'].mean())
data['Ta_m (C)'] = data['Ta_m (C)'].fillna(data['Ta_m (C)'].mean())
data['Ta_l (C)'] = data['Ta_l (C)'].fillna(data['Ta_l (C)'].mean())
data['Operative temperature (C)'] = data['Operative temperature (C)'].fillna(data['Operative temperature (C)'].mean())
data['Radiant temperature (C)'] = data['Radiant temperature (C)'].fillna(data['Radiant temperature (C)'].mean())
data['Globe temperature (C)'] = data['Globe temperature (C)'].fillna(data['Globe temperature (C)'].mean())
data['Tg_h (C)'] = data['Tg_h (C)'].fillna(data['Tg_h (C)'].mean())
data['Tg_m (C)'] = data['Tg_m (C)'].fillna(data['Tg_m (C)'].mean())
data['Tg_l (C)'] = data['Tg_l (C)'].fillna(data['Tg_l (C)'].mean())
data['Relative humidity (%)'] = data['Relative humidity (%)'].fillna(data['Relative humidity (%)'].mean())
data['Air velocity (m/s)'] = data['Air velocity (m/s)'].fillna(data['Air velocity (m/s)'].mean())
data['Velocity_h (m/s)'] = data['Velocity_h (m/s)'].fillna(data['Velocity_h (m/s)'].mean())
data['Velocity_m (m/s)'] = data['Velocity_m (m/s)'].fillna(data['Velocity_m (m/s)'].mean())
data['Velocity_l (m/s)'] = data['Velocity_l (m/s)'].fillna(data['Velocity_l (m/s)'].mean())
data['Blind (curtain)'] = data['Blind (curtain)'].fillna(data['Blind (curtain)'].mode()[0])
data['Fan'] = data['Fan'].fillna(data['Fan'].mode()[0])
data['Window'] = data['Window'].fillna(data['Window'].mode()[0])
data['Door'] = data['Door'].fillna(data['Door'].mode()[0])
data['Heater'] = data['Heater'].fillna(data['Heater'].mode()[0])
data['Outdoor monthly air temperature (C)'] = data['Outdoor monthly air temperature (C)'].fillna(data['Outdoor monthly air temperature (C)'].mean())
```

```python
#clustering

from sklearn.cluster import KMeans                   # K-Means Clustering
from sklearn.cluster import AgglomerativeClustering  # Hierarchical Clustering
from sklearn.cluster import DBSCAN                   # DBSCAN Clustering

from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
n_clusters=len(np.unique(y))
n_clusters
```

```python
plt.scatter(X.iloc[:, 7], X.iloc[:, 18], c=y, cmap='viridis')
plt.title(f'Actual Clusters - Thermal comfort')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
#k means clustering

kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
labels_kmeans = kmeans.labels_
```

```python
# Evaluating the clustering model
silhouette_kmeans = silhouette_score(X, labels_kmeans)
print(silhouette_kmeans)
```

```python
# Plot the clusters
plt.scatter(X.iloc[:, 8], X.iloc[:, 19], c=labels_kmeans, cmap='viridis')
plt.title(f'Kmeans Clustering Results - Thermal Comfort')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
#dbscan clustering

dbscan = DBSCAN(eps=0.6, min_samples=10)
dbscan.fit(X)
labels_dbscan = dbscan.labels_
```

```python
#evaluating clustering model

silhouette_dbscan = silhouette_score(X, labels_dbscan)
print(silhouette_dbscan)
```

```python
# Plot the clusters
plt.scatter(X.iloc[:, 8], X.iloc[:, 19], c=labels_dbscan, cmap='viridis')
plt.title(f'DBSCAN Clustering Results - Thermal comfort')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
#Test train data set splitting

from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier

from sklearn.model_selection import cross_val_score
```

```python
# Define the models for evaluation

models = [
    ('Logistic Regression', LogisticRegression()),
    ('Decision Tree', DecisionTreeClassifier()),
    ('Random Forest', RandomForestClassifier()),
    ('Naive Bayes', GaussianNB()),
    ('Neural Network', MLPClassifier(max_iter=10000)),
]
```

```python
# Evaluate each model using cross-validation
for name, model in models:
    scores = cross_val_score(model, X_train, y_train, cv=5)
    print(f"{name}: Accuracy = {scores.mean()}, Standard Deviation = {scores.std()}")
```
